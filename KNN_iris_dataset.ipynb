# Preparar para a modelagem

```python
X_train = df_train.drop(columns=["target", "target_name"]).values
Y_train = df_train["target"].values
```

```python
Y_train.shape
```

# Modelando - baseline


```python
def single_feature_prediction(petal_length):
  """Preve a espécie de Iris a partir do comprimento da pétala"""
  if petal_length < 2.5:
    return 0
  elif petal_length < 4.7:
    return 1
  else:
    return 2
```

```python
X_train[:, 2]
```

```python
manual_y_predicitons = np.array([single_feature_prediction(val) for val in X_train[:, 2]])
```

```python
manual_model_acuracy = np.mean(manual_y_predicitons == Y_train)
```

```python
print (f"Manual model accuracy: {manual_model_acuracy * 100:.2f}%")
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
model = KNeighborsClassifier(n_neighbors=20)
```

```python
#Separando os dados de teste entre teste e validação
Xt, Xv, Yt, Yv = train_test_split(X_train, Y_train, test_size=0.2)
```

```python
model.fit?
```

```python
model.fit (Xt, Yt) ##treinando
```

```python
Y_pred = model.predict (Xv) #predicted
```

```python
np.mean(Y_pred == Yv)
```

# Tuning

```python
param_grid = {
    'n_neighbors': [5, 10, 20, 30, 50],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}
```

```python
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(
    estimator=KNeighborsClassifier(),  # O modelo a ser otimizado
    param_grid=param_grid,
    scoring='accuracy',
    cv=5
)
```

```python
grid_search.fit(X_train, Y_train)
```

```python
best_params = grid_search.best_params_
print(f"Melhores hiperparâmetros: {best_params}")
```

```python
best_model = grid_search.best_estimator_
test_set_accuracy = best_model.score(X_test, Y_test)
print(f"Accuracy: {test_set_accuracy * 100:.2f}%")
```
